```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import trapz
```

```python
L = 133*1e-3
T = 300
kb = 1.38*1e-23
N = 1
m = 3.66*1e-25
l = N/(2*L)
```

```python
def boltzmann_distr(v,T):
    return np.sqrt(m/(2*np.pi*kb*T))*np.exp(-m*v**2 / (2*kb*T))
```

```python
xs = np.linspace(0,2*L,1000)
def Rate(t):
    rhos = boltzmann_distr((2*L-xs)/t,T)
    integral = trapz(rhos*(2*L-xs)/t,xs)
    return l*integral
```

```python
ts = np.linspace(1*1e-10,10*1e-3,1000)
rs = []
for t in ts:
    r = Rate(t)
    rs.append(r)
plt.plot(ts,rs)
```

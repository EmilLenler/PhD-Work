```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
DATA_NewHV = np.genfromtxt('HV_New_Front_Electrospray.csv', delimiter=';')
#For some reason 0 is set to nan, so fix this
DATA_NewHV[0][0] = 0

print(DATA_NewHV)
Dial_vals = []
RealVoltages = []
for data_pair in DATA_NewHV:
    Dial_vals.append(data_pair[0])
    RealVoltages.append(data_pair[1])

```

```python
plt.rc('font', size = 14)
fig,ax = plt.subplots(figsize=(11.69, 8.27), dpi=100)
ax.scatter(Dial_vals,RealVoltages,)
ax.set_ylabel('Proper Output Voltage / V')
ax.set_xlabel('Dial Position')
ax.set_title('Output voltage vs. dial position\n for new HV supply (FRONT)')
```

```python
#Data clearly seems linear, so let's try to fit to a linear model.

fit_params = np.polyfit(Dial_vals,RealVoltages,1)
print(fit_params)
```

```python
def FittedModel(D):
    return fit_params[0]*D+fit_params[1]

DialSpace = np.linspace(0,6,100)
```

```python
ax.plot(DialSpace,FittedModel(DialSpace),color ='orange', ls = '--')

ax.text(1,-2500,'V = {0}V*d + {1}V'.format(np.round(fit_params[0],1),np.round(fit_params[1],2)))
fig
```

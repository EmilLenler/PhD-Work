```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
R6GDATA = np.loadtxt(r'C:\Users\au581149\PhD-Work\Lab Stuff\ESI\Time of flight\R6G_ToF_28_02_2025.txt',delimiter = ',',skiprows=1)
LYSDATA = np.loadtxt(r'C:\Users\au581149\PhD-Work\Lab Stuff\ESI\Time of flight\LysToF_28_02_2025.txt',delimiter = ',',skiprows = 1)

n_shots = 30/0.05
offset = 11
R6GCounts = R6GDATA[:,0]
R6GTimes = R6GDATA[:,1] * 1e-3 - R6GDATA[0,1] * 1e-3-offset

LYSCounts = LYSDATA[:,0]
LYSTimes = LYSDATA[:,1] * 1e-3 - LYSDATA[0,1] * 1e-3-offset


fig,ax = plt.subplots()
ax.errorbar(R6GTimes,R6GCounts/n_shots,yerr = np.sqrt(R6GCounts)/n_shots,fmt = 'o',markersize = 3,ls = '--',label = 'R6G', color = 'xkcd:dark red',capsize = 3)
ax.errorbar(LYSTimes,LYSCounts/n_shots,yerr = np.sqrt(LYSCounts)/n_shots,fmt = 'o',markersize = 3,ls = '--',label = 'Lys', color = 'xkcd:black',capsize = 3)
ax.set_ylabel('Measured ions per shot')
ax.set_xlabel('Delay / µs')
ax.set_xlim(0,100)

r6gline = 34.5*np.sqrt(84/130)
ax.axvline(r6gline,label = r'm := 130 amu -- Lys(+) - NH3')
ax.axvline(r6gline*np.sqrt(443/130),label = 'm = 443 -- R6G(+)',color = 'xkcd:dark blue')
ax.axvline(r6gline*np.sqrt(147/130),label = 'm = 147 -- Lys(+)',color = 'xkcd:red')
ax.axvline(r6gline*np.sqrt(84/130),label = 'm = 84 -- Lys(+) - (NH3+CO+H2O)',color = 'xkcd:dark blue')


ax.legend()

```

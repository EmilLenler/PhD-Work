```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def a0(q):
    return -0.5*q**2+7/128*q**4-29/2304*q**6+68687/18874368*q**8

def b1(q):
    return 1-q-1/8*q**2+1/64*q**3-1/1535*q**4-11/36864*q**5

def eqOmot(t,q,r0,omega0,omegaRF):
    return r0*(1-q/2*np.cos(omegaRF*t))*np.sin(omega0*t)
```

```python
plt.rcParams.update({'font.size': 22})

qs = np.linspace(0,1.2)
fig,ax = plt.subplots(figsize = (15,10))
ax.fill_between(qs,a0(qs),b1(qs),label = 'Stable region')
ax.plot(qs,a0(qs), color = 'k',linewidth = 3)
ax.plot(qs,b1(qs),color = 'k',linewidth = 3)
ax.set_ylim(np.min(a0(qs)),0)
ax.set_xlim(0,qs[-1])
ax.set_xlabel(r'$q_\rho$')
ax.set_ylabel(r'$a$')
ax.legend()
```

```python
ts = np.linspace(0,5*0.00001,10000)
omega0 = 2*np.pi*50*1e3
omegaRF = 5.2*1e6*2*np.pi
r0 = 5*1e-6

fig,ax = plt.subplots(figsize=(15,10))
ax.plot(ts*1e3,eqOmot(ts,0.1,r0,omega0,omegaRF)*1e6,label ='Solution including micromotion')
ax.plot(ts*1e3,r0*np.sin(omega0*ts)*1e6,label = 'Pseudopotential approximation',linewidth = 3)
ax.legend(loc = 'lower left')
ax.set_xlabel('Time / ms')
ax.set_ylabel(r'Position / $\mu$m')
```

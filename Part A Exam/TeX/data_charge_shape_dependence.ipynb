```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
delay30sec = [0,50,100,200,300,400,500,750,1000,1500,2000]
counts30sec = np.array([2,143,207,193,153,127,96,38,20,14,8])

delay1min = [0,50,100,200,300,400,500,750,1000,1500,2000]
counts1min = np.array([1,180,231,176,170,136,104,56,18,9,5])

delay0 = [0,50,100,200,300,400,500,750,1000,1500,2000]
counts0 = np.array([0,12,27,34,25,34,32,21,18,7,3])
```

```python
fig,ax = plt.subplots()
ax.errorbar(delay30sec,counts30sec,np.sqrt(counts30sec),ls = 'None')
ax.scatter(delay30sec,counts30sec, label = '30sec fill time')

ax.errorbar(delay1min,counts1min,np.sqrt(counts1min),ls = 'None')
ax.scatter(delay1min,counts1min, label = '1 minute fill time')

ax.errorbar(delay0,counts0,np.sqrt(counts0),ls = 'None')
ax.scatter(delay0,counts0,label = 'No fill time')
ax.set_xlabel('Time / s')
ax.set_ylabel('Acc. counts over 20 pulses')

ax.legend()
```

Renormalize the data

```python
def normalize(x):
    return x /np.max(x)
max30 = np.max(counts30sec)
max1m = np.max(counts1min)
max0 = np.max(counts0)

fig,ax = plt.subplots()
ax.errorbar(delay30sec,normalize(counts30sec),np.sqrt(counts30sec)/max30,ls = '--')
ax.scatter(delay30sec,normalize(counts30sec),label = '30 sec fill time')

ax.errorbar(delay1min,normalize(counts1min),np.sqrt(counts1min)/max1m,ls = '--')
ax.scatter(delay1min,normalize(counts1min),label = '1 minute fill time')

ax.errorbar(delay0,normalize(counts0),np.sqrt(counts0)/max0,ls = '--')
ax.scatter(delay0,normalize(counts0), label = 'No fill time')
ax.set_xlabel('Time / s')
ax.set_ylabel('Signal normalized to peak value')
ax.legend()
```

```python
fig,ax = plt.subplots()
ax.errorbar(delay30sec,counts30sec/np.sum(counts30sec),np.sqrt(counts30sec)/np.sum(counts30sec),ls = '--')
ax.scatter(delay30sec,(counts30sec)/np.sum(counts30sec),label = '30 sec fill time')

ax.errorbar(delay1min,(counts1min)/np.sum(counts1min),np.sqrt(counts1min)/np.sum(counts1min),ls = '--')
ax.scatter(delay1min,(counts1min)/np.sum(counts1min),label = '1 min fill time')

ax.errorbar(delay0,counts0/np.sum(counts0),np.sqrt(counts0)/np.sum(counts0),ls = '--')
ax.scatter(delay0,(counts0)/np.sum(counts0), label = 'No fill time')
ax.set_xlabel('Time / s')
ax.set_ylabel('Signal normalized to sum signal')
ax.legend()
```

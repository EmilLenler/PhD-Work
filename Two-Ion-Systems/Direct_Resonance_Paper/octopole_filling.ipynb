```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fill_times = np.array([1,2,3,4,5,6,7,10,15,20,25])

counts = np.array([1432,1920,2266,2434,2725,2868,2903,3157,3995,4166,4266])

current_before = np.array([600,800,800,750,750,800,650,700,650,650,700])

current_after = np.array([600,850,800,800,750,650,700,650,600,600,700])

mean_current = current_after+current_before/2


normalized_to = mean_current[0]
norm_counts = counts * (normalized_to/mean_current)
```

```python
fig,ax = plt.subplots(figsize =(12,8))
ax.scatter(fill_times,counts)
ax.errorbar(fill_times,counts,np.sqrt(counts))
ax.set_xlabel('Octopole Fill Time (s)')
ax.set_ylabel('Accumulated counts over 20 repeats')
ax.set_title('Unadjusted Counts')
ax.set_ylim(0,5000)
```

```python

fig,ax = plt.subplots(figsize =(12,8))
ax.scatter(fill_times,norm_counts)
ax.errorbar(fill_times,norm_counts,np.sqrt(counts)*(normalized_to/mean_current))
ax.set_xlabel('Octopole Fill Time (s)')
ax.set_ylabel('Accumulated counts over 20 repeats')
ax.set_title('Counts adjusted for mean current at time of measurement')
ax.set_ylim(0,5000)
```
